```python
# Target variable
y = df['Churn']

# Features (everything except Churn)
X = df.drop('Churn', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 80/20 split
    stratify=y,           # IMPORTANT
    random_state=42       # for reproducibility
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit ONLY on training data
X_train = scaler.fit_transform(X_train)

# Apply same transformation to test data
X_test = scaler.transform(X_test)
```

```python
X_train['international_plan'] = X_train['international_plan'].map({'Yes': 1, 'No': 0})
X_test['international_plan'] = X_test['international_plan'].map({'Yes': 1, 'No': 0})

X_train['voice_mail_plan'] = X_train['voice_mail_plan'].map({'Yes': 1, 'No': 0})
X_test['voice_mail_plan'] = X_test['voice_mail_plan'].map({'Yes': 1, 'No': 0})
```
