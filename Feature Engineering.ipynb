```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:/Users/ashuy/OneDrive/Desktop/FINAL YEAR PROJECT/code/Intelligent-Customer-Churn-Prediction-System/cleaned_dataset.csv")
```

```python
#Create tenure groups (using account_length)
df['tenure_group'] = pd.cut(df['account_length'],
                            bins=[0,50,100,150,200,300],
                            labels=['0-50','50-100','100-150','150-200','200+'])
```

```python
#Create average revenue per month
df['avg_revenue_per_month'] = (
    df['total_day_charge'] +
    df['total_eve_charge'] +
    df['total_night_charge'] +
    df['total_intl_charge']
)
```

```python
#Create total spending feature
df['total_spending'] = (
    df['total_day_charge'] +
    df['total_eve_charge'] +
    df['total_night_charge'] +
    df['total_intl_charge']
) * df['account_length']
```

```python
#Convert categorical columns
#label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['international_plan'] = le.fit_transform(df['international_plan'])
df['voice_mail_plan'] = le.fit_transform(df['voice_mail_plan'])
```

```python
#One-Hot Encoding
df = pd.get_dummies(df, columns=['state','tenure_group'], drop_first=True)
```

```python
#Scale numeric features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numeric_cols = df.select_dtypes(include=['int64','float64']).columns

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
#Check imbalance ratio
df['Churn'].value_counts()
```

```python
df = df.drop(columns=['customer_id'])
```

```python
#Convert categorical columns to numeric
df['international_plan'] = df['international_plan'].map({'yes':1, 'no':0})
df['voice_mail_plan'] = df['voice_mail_plan'].map({'yes':1, 'no':0})
```
